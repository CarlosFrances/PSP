```python
''' 
    Operadores
'''

print ('Operadores matemáticos: \n')
print ('Suma, operador +: 2 + 2 = ', 2 + 2)
print ('Resta, operador -: 4 - 2 = ', 4 - 2)
print ('Multiplicación, operador *: 3 * 3 = ', 3 * 3)
print ('División, operador /: 6 / 2 = ', 6 / 2)
print ('Exponencial, operador /: 6 ** 2 = ', 6 ** 2)
print ('Cociente, operador //: 9 // 2 = ', 9 // 2)
print ('Resto, operador %: 9 % 2 = ', 9 % 2)
print ('')

'''
    Operadores relacionales
'''
print ('Operadores relacionales: \n')
print ('Igual que, operador ==: 2 == 2 = ', 2 == 2)
print ('Distinto que, operador !=: 4 != 2 = ', 4 != 2)
print ('Menor que, operador <: 3 < 3 = ', 3 < 3)
print ('Mayor que, operador >: 6 > 2 = ', 6 > 2)
print ('Menor o igual que, operador <=: 6 <= 2 = ', 6 <= 2)
print ('Mayor o igual que, operador >=: 9 >= 2 = ', 9 >= 2)
print ('')

'''
    Operadores lógicos
'''
print ('Operadores lógicos: \n')
print ('y, operador and: 2==2 and 3==3', 2==2 and 3==3)
print ('o, operador or: 2==2 or 3==3', 2==2 or 3==3)
```

### EJERCICIO: Operadores simples

**Diseña un programa que resuelva la siguiente situación**: 

Tres camiones van a distribuir calabazas en cajas de 8 kilos. ¿Cuántas cajas/kilos son necesarias para cada camión, si cada camión transporta las siguientes cantidades:
    - Camión 1: 3464 kilos.
    - Camión 2: 664 kilos menos que el camión 1.
    - Camión 3: 220 cajas

```python
camion1 = 3464
camion2 = camion1-664
camion3 = 220*8

print("Camión 1: ",camion1)
print("Camión 2:",camion2)
print("Camión 3:",camion3)

camion1//=8
camion2//=8
camion3//=8

print("Camión 1: ",camion1)
print("Camión 2:",camion2)
print("Camión 3:",camion3)
```
